```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Using MatplotLib to Chart a DataFrame

```python
# Load in csv
all_df = pd.read_csv("Resources/person.csv")
all_df.head()
all_df.columns
```

```python
def_df = pd.read_csv("Resources/body_type.csv")
def_df.head()
# all_df.columns
```

```python
# Get body_type data
body_type_data = all_df["BODY_TYP"].value_counts()
body_type_data.sort_values(ascending=False)
first_40_types = body_type_data.nlargest(40)

first_40_types.values

body_type_df = pd.DataFrame({'BODY_TYPE' : first_40_types.index, 
    'accident_total' : first_40_types.values
})

body_type_df

final_data = pd.merge(body_type_df, def_df, on = 'BODY_TYPE')
final_data
```

```python
# Configure plot and ticks
first_40_types.plot(kind="bar", facecolor="red")
```

```python
# x_axis = np.arange(0, 40)
# tick_locations = []
# for x in x_axis:
#     tick_locations.append(x)

plt.title("Body Types Accidents")
plt.xlabel("Body Type")
plt.ylabel("Number of Accedents")

# plt.xlim(-0.75, len(gyms)-.25)
# plt.ylim(0, max(members) + 5)

plt.bar(final_data['Definition'], final_data['accident_total'] , facecolor="red", alpha=0.75, align="center")
# plt.xticks(tick_locations, gyms)
plt.show()
```
